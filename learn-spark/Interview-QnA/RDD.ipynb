# Resilient Distributed Dataset (RDD)

### <u>What is RDD?</u>

RDDs is the way how Spark achieves "_Fault Tolerance_" and "_Efficiency / Speed_"

RDD - Resilient Distributed Dataset is the basic unit of abstraction in Spark. RDD is :-- 

- Resilient and Immutable collection of elements / records
- Distributed collection of elements / records 
- Partitioned collection of elements / records

Resilient \>\> RDDs are resilient to failures since spark keeps a lineage of operations perfomed on dataset. In case of failure of any particular node in cluster Spark can recover the partition by executing same set of commands as it did last time to arrive at current partition.

Distributed Dataset \>\> Data is distributed across nodes in cluster and can be processed in parallel.

  

Immutable \>\> elements in RDDs cannot be changed randomly. The operations performed on RDD is applied to all the partitions at same time. To have a different operation performed on any RDD one has to create a seperate RDD with desired operations applied to it. 

\\>\> with help of RDD spark keep track of all the operations performed on a dataset.

\\>\> All RDDs do not persist in memory by default. One can use "_.cache()"_ function to persist RDD in memory.
